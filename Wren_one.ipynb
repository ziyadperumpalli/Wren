```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data= pd.read_csv("datasheet.csv")
data
```

```python
#Generate Random Heights
data["Height"]=np.random.randint(150,181,size=len(data))
```

```python
# to remove any rows containing missing (NaN) values from the DataFrame.
data.dropna(inplace=True)
```

```python
data
```

### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_count=data["Team"].value_counts()
percentage_splitting=np.round((team_count/len(data))*100,2)
percentage_splitting
```

### 2.Segregate the employees w.r.t different positions.

```python
count=data["Position"].value_counts()
df1=pd.DataFrame(count)
df1.columns = ["Count"]
df1
```

### 3.Find from which age group most of the employees belong to.

```python
x=data["Age"]
plt.hist(x)
plt.xlabel("Age")
plt.xlabel("Number of Employees")
```

### 4.Find out under which team and position, spending in terms of salary is high.

```python
data[['Team','Position','Salary']].sort_values('Salary',ascending=False).head(1)
```

### 5. Find if there is any correlation between age and salary , represent it visually.

```python
df2=data[["Age","Salary"]]
corr=df2.corr()
sns.heatmap(corr,annot=True, cmap="viridis")
plt.title("Correlation between age and salary")

```
